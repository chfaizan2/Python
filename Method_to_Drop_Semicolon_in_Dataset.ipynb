```python
df.dropna(inplace=True)
```

```python
def to_number(x):
  x = x.replace(",", "")
  x = int(x)
  return x
```

```python
df.Total_Recovered = df.Total_Recovered.apply(to_number)
df.Active_Cases = df.Active_Cases.apply(to_number)
df["Total Cases"] = df["Total Cases"].apply(to_number)
df.Total_Deaths  = df.Total_Deaths.apply(to_number)
df["Tests/1M_pop"]  = df["Tests/1M_pop"].apply(to_number)
df.Total_Tests  = df.Total_Tests.apply(to_number)
df["Deaths/1M_pop"]  = df["Deaths/1M_pop"].apply(to_number)
df["Tot_Cases/1M_pop"] = df["Tot_Cases/1M_pop"].apply(to_number)
df.Population   = df.Population .apply(to_number)

```

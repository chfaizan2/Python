### fillna

```python
import pandas as pd
x= pd.read_csv("/CarPricesPrediction.csv")
print(x)
```

```python
x.fillna(2)
```

```python
x.fillna({"Price":23,"Model":"toyo"})
```

```python
x.fillna(method="ffill")
```

```python
x.fillna(method="bfill")
```

```python
x.fillna(method="ffill", axis=1)
```

```python
x.fillna(method="ffill", axis=0)
```

```python
x.fillna(method="ffill", limit=1)
```

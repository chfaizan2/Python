```python
import pandas as pd
df= pd.read_csv("/content/pakwheels_used_car_data_v02.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
body_mode= df.body.mode()
body_mode[0]
```

```python
df["body"]= df["body"].fillna(body_mode[0])
```

```python
year_mean= df.year.mean()
year_mean
```

```python
df["year"]= df["year"].fillna(year_mean)
```

```python
engine_mean= df.engine.mean()
engine_mean
```

```python
df["engine"]= df["engine"].fillna(engine_mean)
```

```python
fuel_mode= df.fuel.mode()
fuel_mode[0]
```

```python
df["fuel"]= df["fuel"].fillna(fuel_mode[0])
```

```python
color_mode= df.color.mode()
color_mode[0]
```

```python
df["color"]= df["color"].fillna(color_mode[0])
```

```python
price_mean= df.price.mean()
price_mean
```

```python
df["price"]= df["price"].fillna(price_mean)
```

```python
df= df.drop(["assembly","registered"], axis=1)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder= LabelEncoder()
```

```python
df["body"]= label_encoder.fit_transform(df["body"])
```

```python
df["model"]= label_encoder.fit_transform(df["model"])
```

```python
df["transmission"]= label_encoder.fit_transform(df["transmission"])
```

```python
df["fuel"]= label_encoder.fit_transform(df["fuel"])
```

```python
df["color"]= label_encoder.fit_transform(df["color"])
```
